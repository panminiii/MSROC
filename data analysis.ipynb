```python
import matplotlib.pyplot as plt
```

## 模型的 Step 范围

```python
loss_data = []
for i in range(1, 11):
    with open(f"./models/{i}/loss_train.txt") as f:
        loss_data.append(eval(f.read()))

tal = 0
for i, d in enumerate(loss_data):
    tal += len(d)
    print(f"Model{i} - to", tal)
```

```python
loss_data = []
acc_data = []

for i in range(1, 11):
    with open(f"./models/{i}/loss_train.txt") as f:
        loss_data.extend(eval(f.read()))
    with open(f"./models/{i}/acc_train.txt") as f:
        acc_data.extend(eval(f.read()))
```

## 训练Loss | 发现 15000 - 20000 之间的梯度爆炸

```python
plt.figure()
plt.plot(range(len(loss_data)), loss_data)
plt.show()
```

## 训练 Accuracy |

```python
plt.figure()
plt.plot(range(len(acc_data)), acc_data)
plt.show()
```
